# Visual Analytics - Seaborn

Seaborn is python data visualization library that is based on matplotlib.

## Why is Data Visualization Important? 

### Data visualization is important because it:

1. **Makes data easier to understand**: Visualizations present data in a visual format, helping us understand complex patterns, trends, and relationships more easily.

2. **Enables data exploration**: Visualizations allow us to interact with data, explore different perspectives, and uncover insights that may not be apparent in raw data.

3. **Simplifies communication**: Visualizations simplify the communication of data to others, making it easier for them to understand and engage with the information.

4. **Supports storytelling and persuasion**: Visualizations help tell data-driven stories and make persuasive arguments by presenting information in a compelling and memorable way.

5. **Identifies patterns and outliers**: Visualizations highlight patterns, trends, and outliers, helping us identify anomalies and important insights in the data.

6. **Aids decision-making**: Visualizations enable us to analyze and compare options, assess outcomes, and make informed decisions based on data.

7. **Promotes collaboration**: Visualizations facilitate discussions, align different perspectives, and promote teamwork by providing a shared understanding of data.

8. **Improves data quality assessment**: Visualizations help us spot data quality issues like missing values or inconsistencies, leading to better data cleaning and analysis.

In essence, data visualization simplifies data understanding, aids decision-making, and enhances communication by presenting data visually in a way that engages and empowers both analysts and stakeholders.

## Categorical Plot Examples

### barplot

The x-axis of the bar plot represents the different categories or levels of the 'sex' variable, which could be something like 'male' and 'female'. Each category is associated with a bar on the plot.

The y-axis represents the corresponding values of the 'total_bill' variable for each category. The height of each bar indicates the average value of 'total_bill' for that particular category.

The bar plot allows you to compare the average 'total_bill' between different categories of 'sex'. It provides a visual representation of the distribution and comparison of 'total_bill' values for each gender category in the dataset.

The black line in the middle of the bars represents the confidence interval:
1. It indicates the **range of uncertainty** around the **estimated mean value** for each bar in the bar plot. 
2. It helps assess the **variability and reliability** of the data. 
3. To remove the black line, you can use the parameter **ci=None** when creating the bar plot.

Black line represents confidence interval

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

 The line **<AxesSubplot:xlabel='sex', ylabel='total_bill'>** represents a single plot with the x-axis labeled as "sex" and the y-axis labeled as "total_bill". It confirms that the plot has been created and provides information about its characteristics.

This line is a way for the plotting library to indicate the existence of the plot and provide details about its axes labels. It helps ensure that your plots are generated correctly and allows you to identify specific plots within a larger visualization.

### countplot
Used to count samples.

A count plot is a type of bar plot that displays the frequency or count of occurrences of each category in a categorical variable

```python
sns.countplot(x='sex',data=df)
```

### boxplot
box plots show the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range. They are used to show the distribution of categorical data

```python
sns.boxplot(x="day", y="total_bill", data=df)
```

#### Hue
The hue argument can be used to split the data based on a 3rd feature. This causes each exiting box plot to be split into two corressponding to the 3rd feature.

```python
sns.boxplot(x="day", y="total_bill", hue="smoker",data=df)
```

### violinplot
violin plots are used for similar reasons to box plots. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.

```python
sns.violinplot(x="day", y="total_bill", data=df)
```

#### Hue
The hue argument can similarly be used with violin plots but violin plots also take an additional argument called `split` which allows each half of the violin plot to correspond to a value of the new feature.

```python
sns.violinplot(x="day", y="total_bill", data=df,hue='sex',split=True)
```

### stripplot
The stripplot will draw a scatterplot where one variable is categorical. A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

```python
sns.stripplot(x="day", y="total_bill", data=df)
```

#### jitter
The `jitter` argument allows us to scatter the plots a bit so we can see how dense the data is.

```python
sns.stripplot(x="day", y="total_bill", data=df, jitter=True)
```

#### hue
The `hue` argument allows us to color the plots based on a 3rd argument. Here the additional `split` argument will split the stripplot into two separate plots.

##### w/o split

```python
sns.stripplot(x="day", y="total_bill", data=df, jitter=True, hue='sex',dodge=False)
```

##### w/ split

```python
sns.stripplot(x="day", y="total_bill", data=df, jitter=True, hue='sex',dodge=True)
```

### swarmplot

The swarmplot is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they don’t overlap. This gives a better representation of the distribution of values, although it does not scale as well to large numbers of observations (both in terms of the ability to show all the points and in terms of the computation needed to arrange them).

1. Better distribution than stripplot()
2. No overlapping

```python
sns.swarmplot(x="day", y="total_bill", data=df)
```

```python
sns.swarmplot(x="day", y="total_bill", hue='sex', data=df, dodge=True, size=4)
```

### catplot

catplot is the most general form of a categorical plot. It can take in a **kind** parameter to adjust the plot type:

The catplot function simplifies the process of creating categorical plots by providing a unified interface and allowing you to specify the type of plot using the kind parameter.

In this code, a bar plot is created with 'sex' categories on the x-axis and 'total_bill' values on the y-axis. The bar plot represents the average total bill for each category of 'sex', allowing you to compare the values between different genders easily.

```python
sns.catplot(x='sex',y='total_bill',data=df,kind='bar')
```

When you create a FacetGrid object and print it in the notebook, you see the representation **<seaborn.axisgrid.FacetGrid at 0x7fd9e24db850>**, which indicates that the PairGrid object has been created and provides the memory address where it is stored in the computer's memory.

 The line **<seaborn.axisgrid.FacetGrid at 0x7fd9e24db850>** is just a representation that appears when you create a Seaborn FacetGrid object and print it. It confirms that the object has been successfully created.

## Distribution Plots

### distplot

```python
sns.histplot(df['total_bill'], kde=True, stat="density", linewidth=0)
```


1. stat="density" represents the histogram as a density plot, showing the probability density.
2. linewidth=0 removes the lines outlining the histogram bars, making them invisible.

Other Options for stat:

`stat` must be one of ['count', 'frequency', 'density', 'probability', 'proportion', 'percent'], but 'sum' was passed.

#### w/o kde

```python
sns.histplot(df['total_bill'], kde=False, stat="density", linewidth=0, bins=30)
```

### jointplot

jointplot() allows you to basically match up two distplots for bivariate data. With your choice of what **kind** parameter to compare with: 
* “scatter”: Shows a scatter plot of the data points.
* “reg”: Adds a regression line to the scatter plot.
* “resid”: Displays the residuals of the regression line.
* “kde”: Creates a 2D kernel density plot.
* “hex”: Generates a hexbin plot to visualize density.

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

The resulting joint plot will display a hexbin plot, which is a 2D histogram, showing the relationship between 'total_bill' and 'tip' variables. 

The hexagons represent bins, and the color intensity or shading within each hexagon indicates the **density or frequency** of data points falling within that bin. 

This plot helps visualize the **distribution and density** of data points and identify areas of **high or low concentration**.

### pairplot

pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns) and supports a color hue argument (for categorical columns). 

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

### rugplot

rugplots are actually a very simple concept, they just draw a dash mark for every point on a univariate distribution. They are the building block of a KDE plot:

```python
sns.rugplot(df['total_bill'])
```

### kdeplot
kdeplots or Kernal Density Estimation plots replace every single observation with a Gaussian (Normal) distriubtion centered around that value.

```python
sns.kdeplot(df['total_bill'])
sns.rugplot(df['total_bill'])
```

## Regression plots

### lmplot
Draw a linear regression plot

A regression plot is a visualization that shows the relationship between two variables using a scatter plot with a fitted line. It helps understand if there is a linear trend and the strength of the relationship between the variables.

Regression analysis models the **relationship between a dependent variable and independent variables.** 

It helps understand how changes in the independent variables are associated with changes in the dependent variable, allowing for prediction and analysis of relationships between variables.

```python
sns.lmplot(x='total_bill',y='tip',data=df)
```

#### Using hue

```python
sns.lmplot(x='total_bill',y='tip',data=df, hue='sex')
```

#### Using col

```python
sns.lmplot(x='total_bill',y='tip',data=df,col='sex')
```

## Matrix Plots
Matrix plots allow you to plot data as color-encoded matrices.

### heatmap

1. `df.corr()`: Calculates the **correlation matrix** of the DataFrame df. The correlation matrix shows the pairwise correlation between variables, indicating how closely related they are.

2. `sns.heatmap()`: Creates a heatmap plot using Seaborn. The heatmap represents the **correlation matrix** as a color-coded grid.

```python
sns.heatmap(df.corr())
```

#### w/ annot (annotation)

When `annot=True`, each cell in the heatmap will display the numeric value representing the correlation between the corresponding pair of variables. 

These annotations provide additional information about the **magnitude or strength of the correlation** between the variables.

```python
sns.heatmap(df.corr(), annot=True)
```

```python
flights = sns.load_dataset('flights')
pvflights = flights.pivot_table(values='passengers',index='month',columns='year')
sns.heatmap(pvflights)
```

#### w/ separation using linecolor and linewidths

```python
sns.heatmap(pvflights,cmap='magma',linecolor='white',linewidths=1)
```

### clustermap

The clustermap uses hierarchal clustering to produce a clustered version of the heatmap.

```python
sns.clustermap(pvflights)
```

#### w/ scaling from 0-1 using standard_scale

```python
sns.clustermap(pvflights,standard_scale=1)
```
