### source column

```python
df.source.unique()
```

```python
# change "METRO" from source column from the dataframe
df['source'] = df['source'].replace('METRO', 'Company A')
```

```python
df.source.unique()
```

### Banner

```python
df.Banner.unique()
```

```python
df['Banner'] = df['Banner'].replace('METRO ONTARIO', 'Company A ONTARIO')
df['Banner'] = df['Banner'].replace('METRO QUEBEC', 'Company A QUEBEC')
df['Banner'] = df['Banner'].replace('SUPERC', 'Company A Super')
```

```python
df.Banner.unique()
```

```python
df.SRC_BAN_POS.unique()
```

```python
df['SRC_BAN_POS'] = df['SRC_BAN_POS'].str.replace('METRO#ALL#', '')
df.head()
```

```python
df.Plan_Banner.unique()
```

```python
df['Plan_Banner'] = df['Plan_Banner'].replace('METRO', 'Company A')
df['Plan_Banner'] = df['Plan_Banner'].replace('SUPER C', 'Company A Sup')
```
