### Significance of vectorized coding:

Writing loops to handle different scenarios in our algorithms might seem the most straightforward solution; but it is not the efficient one. In the next few cells, we will try to do an element-wise vector multiplication on A and B vectors. Each vector has 10 million elements. First, we will multiply each element of A with the corresponding element of B and store the result in C using a for loop. Then we try to implement it by vectorized operation only as `A * B`. Let's see what happens!

```python
A = np.arange(0,10000000,1)
```

create an array A with numbers starting from 0 up to (but not including) 10000000, incrementing by 1 in each step.

```python
A
```

```python
B = np.random.random(10000000)
C1 = np.zeros(10000000)
C2 = np.zeros(10000000)
```

1. np: This is an alias for the NumPy library. It is used to access NumPy functions and methods.

2. np.random: This is a module in NumPy that provides functions for generating random numbers and random arrays.

3. np.random.random(10000000): This calls the random function from the np.random module to generate an array of random numbers.

4. random: This function generates random numbers from a uniform distribution, which means that each value has an equal probability of being selected.
10000000: This argument specifies the size of the array to be generated. In this case, the array will have 10,000,000 elements.


```python
import time

start=time.time()

for i in range (0,10000000):
    C1[i]=A[i]*B[i]

elapsed_time1 = time.time()-start

print("For loop Elapsed time: ", elapsed_time1)
```

```python
start=time.time()

C2 = A * B

elapsed_time2 = time.time()-start

print("Vectorized Elapsed time: ", elapsed_time2)
```

```python
Error=np.sum(abs(C2-C1))
print("Error=", Error)
print("Speedup=", elapsed_time1/elapsed_time2)
```

As expected the vectoried computation was faster, but how much? Like $2\times$? or wait... $400\times$!! We should definitely try to avoid loops as much as possible!

You are given the following tasks to perform using NumPy:

1. Create a 3 by 4 matrix using NumPy with random integers in the range 1 to 100.
2. Transpose the matrix to interchange rows and columns.
3. Calculate the mean, minimum, maximum, and sum of the elements in the matrix.

Your task is to write a Python program to accomplish these tasks and display the original matrix, the transposed matrix, and the calculated mean, minimum, maximum, and sum of the matrix.

### Answer

```python
import numpy as np

# Task 1: Create a 3 by 4 matrix with random integers in the range 1 to 100
matrix = np.random.randint(1, 101, size=(3, 4))

# Task 2: Transpose the matrix
transposed_matrix = matrix.T

# Task 3: Calculate the mean, minimum, maximum, and sum of the matrix
mean_value = np.mean(matrix)
min_value = np.min(matrix)
max_value = np.max(matrix)
sum_value = np.sum(matrix)

# Display the original matrix
print("Original Matrix:")
print(matrix)

# Display the transposed matrix
print("\nTransposed Matrix:")
print(transposed_matrix)

# Display the calculated values
print("\nMean:", mean_value)
print("Minimum:", min_value)
print("Maximum:", max_value)
print("Sum:", sum_value)
```

### 2.1.2 Seaborn vs. Matplotlib

Seaborn and Matplotlib are both popular Python libraries used for data visualization, but they serve different purposes and have distinct features. Here's an overview of the key differences, strengths, and weaknesses of each:

1. **Matplotlib**:

- **Purpose:** Matplotlib is a powerful and flexible 2D plotting library that provides a wide variety of plot types and customization options. It is the foundation upon which many other visualization libraries are built.


- **Strengths**:
    1. Comprehensive: Matplotlib offers extensive control over plot elements, allowing users to create highly customized visualizations.
    2. Wide Range of Plot Types: It supports various plot types, such as line plots, scatter plots, bar plots, histograms, pie charts, etc.
    3. Publication Quality: Matplotlib provides precise control over plot aesthetics, making it ideal for creating publication-quality figures.
    4. Low-level Control: It allows fine-grained control over each element of the plot.


- **Weaknesses**:
    1. Verbosity: Creating complex visualizations can lead to verbose code, requiring more lines of code compared to other libraries.
    2. Limited Statistical Plotting: Matplotlib does not offer specialized statistical plotting functions like Seaborn.


2. **Seaborn**:

- **Purpose**: Seaborn is built on top of Matplotlib and provides a higher-level interface for creating attractive statistical visualizations. It is specifically designed for data analysis and exploration.


- **Strengths**:
    1. Simplified Syntax: Seaborn's API is simpler and more intuitive, allowing users to create complex plots with less code.
    2. Statistical Plotting: Seaborn offers numerous statistical plotting functions that are not available in Matplotlib, making it easier to visualize relationships and patterns in data.
    3. Attractive Defaults: Seaborn's default styles and color palettes result in visually appealing plots without much customization.
    4. Works Well with Pandas: Seaborn seamlessly integrates with Pandas DataFrames, making it convenient for data analysis workflows.


- **Weaknesses**:
    1. Limited Customization: While Seaborn's default styles are aesthetically pleasing, customizing the plots beyond these defaults can be more challenging than in Matplotlib.
    2. Fewer Plot Types: Seaborn offers a range of statistical plots, but it may not cover the breadth of plot types available in Matplotlib.


In summary, if you need fine-grained control over plot customization or a wide range of plot types, Matplotlib is a solid choice. However, if you prioritize simplicity, statistical visualization, and visually appealing default styles, Seaborn may be the better option. Often, a combination of both libraries can be used effectively to leverage the strengths of each when creating data visualizations.

#### 2.2.4 boxplot
box plots show the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range. They are used to show the distribution of categorical data


```python
sns.boxplot(x="day", y="total_bill", data=df)
```

1. The Box: The box in the box plot represents the interquartile range (IQR) of the data, which is the middle 50% of the dataset. It is a rectangle that spans from the first quartile (25th percentile or Q1) to the third quartile (75th percentile or Q3). The height of the box represents the IQR. The line inside the box represents the median (50th percentile or Q2) of the data.

2. The Whiskers: The whiskers extend from the edges of the box to the minimum and maximum values within a certain range. The range is determined by the "whisker length" parameter. By default, the whiskers extend to the minimum and maximum values within 1.5 times the IQR from Q1 and Q3, respectively. Data points outside this range are considered potential outliers and are plotted as individual points.

3. Outliers: Data points that fall outside the whiskers' range are plotted as individual points and are considered potential outliers. Outliers are points that deviate significantly from the rest of the data.

4. The Line: The line (usually horizontal) within the box represents the median (Q2) of the data. The median is a measure of the central tendency of the data and divides the data into two equal halves.

- Pros: Reveals the relationship between the whole and its parts, allows easy comparison of categories, and reveals patterns within each category.
- Cons: Challenging to compare absolute values and requires careful consideration of category order.

Regular Bar plots:
- Compares values across different categories or subgroups.
- Pros: Allows direct comparison of values between categories and effective in showing differences and similarities.
- Cons: Can become crowded with many categories and may struggle to represent multiple variables.

#### 2.3.9 Titles, ticks and ticklabels
`y2 = [sum(x < 0 for x in y1), sum(x >= 0 for x in y1)]`



This code uses generator expressions and the sum() function to calculate the counts of elements in y1 that satisfy the conditions x < 0 and x >= 0, respectively, and stores these counts in the y2 list.

`y1 = np.random.normal(-0.07, 0.05, 100)` generates a NumPy array y1 containing 100 random numbers drawn from a normal distribution.

```python
# Create data
y1 = np.random.normal(-0.07, 0.05, 100)
y2 = [sum(x < 0 for x in y1), sum(x >= 0 for x in y1)]

# Create subplot
fig, axs = plt.subplots(1, # the number of rows in the subplot grid
                        2, # the number of columns in the subplot grid
                        figsize=(12, 5.5))

# Plot data
axs[0].plot(y1)
axs[1].bar([0, 1], # x-coordinates of the bars. The values [0, 1] indicate the positions of the bars on the x-axis.
           y2, #  the heights of the bars
           0.8, # width of the bars
           color=["r", "g"], # the colors of the bars
           alpha=0.5 # the transparency of the bars
          )

# Set titles for subplots
axs[0].set_title("Unordered sequence plot")
axs[1].set_title("Values count relative to zero")

# Set title for entire figure
fig.suptitle("Sequence of random values", size=14, weight="bold")

# Remove xticks from the first plot
# xticks( ticks ) sets the x-axis tick values, which are the locations along the x-axis where the tick marks appear
axs[0].set_xticks([])

# Add xticks for the second plot and rename them
axs[1].set_xticks([0, 1])
axs[1].set_xticklabels(["below 0", "greater or equal 0"])
```
